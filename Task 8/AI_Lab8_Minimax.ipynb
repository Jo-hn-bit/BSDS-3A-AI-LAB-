# AI Lab 8 â€” Minimax (quick)

Enter scores as comma separated values (length should be power of 2). It prints the optimal value.


```
import math

def minimax(depth, idx, is_max, scores, target_depth):
    if depth == target_depth:
        return scores[idx]
    if is_max:
        return max(
            minimax(depth+1, idx*2, False, scores, target_depth),
            minimax(depth+1, idx*2+1, False, scores, target_depth)
        )
    else:
        return min(
            minimax(depth+1, idx*2, True, scores, target_depth),
            minimax(depth+1, idx*2+1, True, scores, target_depth)
        )

raw = input("Scores (e.g. 3,5,2,9,3,5,2,9): ").strip()
if not raw:
    raw = "3,5,2,9,3,5,2,9"
scores = []
for x in raw.split(","):
    x = x.strip()
    if x.lstrip("-").isdigit():
        scores.append(int(x))
if not scores:
    scores = [3,5,2,9,3,5,2,9]

treeDepth = int(math.log(len(scores), 2))
print("Optimal value:", minimax(0, 0, True, scores, treeDepth))
```

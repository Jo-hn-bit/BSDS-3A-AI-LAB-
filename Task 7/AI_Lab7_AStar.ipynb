Implements A* like the lab. Enter start/goal and optionally tweak heuristic values.


```
def a_star(adj, h, start, goal):
    open_set = set([start])
    closed_set = set()
    g = {start: 0}
    parent = {start: start}

    while open_set:
        n = None
        for v in open_set:
            if n is None or g.get(v, 10**9) + h.get(v, 10**6) < g.get(n, 10**9) + h.get(n, 10**6):
                n = v
        if n is None:
            print("No path.")
            return []

        if n == goal:
            path = []
            while parent[n] != n:
                path.append(n)
                n = parent[n]
            path.append(start)
            path.reverse()
            print("Path:", path)
            return path

        for (m, w) in adj.get(n, []):
            if m not in open_set and m not in closed_set:
                open_set.add(m)
                parent[m] = n
                g[m] = g[n] + w
            else:
                new_g = g[n] + w
                if g.get(m, 10**9) > new_g:
                    g[m] = new_g
                    parent[m] = n
                    if m in closed_set:
                        closed_set.remove(m)
                        open_set.add(m)
        open_set.remove(n)
        closed_set.add(n)
    print("Path does not exist.")
    return []

adj = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)],
    'D': []
}
h = {'A': 1, 'B': 1, 'C': 1, 'D': 1}

print("A* demo")
s = input("Start (default A): ").strip() or "A"
t = input("Goal (default D): ").strip() or "D"
opt = (input("Change heuristic? (y/n, default n): ").strip().lower() or "n")
if opt == "y":
    for k in list(h.keys()):
        try:
            h[k] = int(input(f"h({k}) = ").strip())
        except:
            pass

path = a_star(adj, h, s, t)
print("Final path:", path)
```

```
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G','H'],
    'E': [],
    'F': ['I','K'],
    'G': [],
    'H': ['L'],
    'I': [],
    'K': ['M'],
    'L': [],
    'M': []
}

def dls(curr, goal, level, max_level, path):
    path.append(curr)
    if curr == goal:
        return True
    if level == max_level:
        path.pop()
        return False
    for child in graph.get(curr, []):
        if dls(child, goal, level+1, max_level, path):
            return True
    if path and path[-1] != goal:
        path.pop()
    return False

def bfs_without_queue(start, goal):
    visited = set()
    arr = [start]
    head = 0
    parent = {start: None}
    order = []
    while head < len(arr):
        node = arr[head]; head += 1
        if node in visited:
            continue
        visited.add(node)
        order.append(node)
        if node == goal:
            p = []
            x = goal
            while x is not None:
                p.append(x)
                x = parent.get(x)
            p.reverse()
            return order, p
        for nb in graph.get(node, []):
            if nb not in visited and nb not in arr:
                arr.append(nb)
                if nb not in parent:
                    parent[nb] = node
    return order, []

class QNode:
    def __init__(self, value):
        self.value = value

def bfs_with_queue(start, goal):
    from collections import deque
    q = deque([QNode(start)])
    visited = set()
    parent = {start: None}
    order = []
    while q:
        qn = q.popleft()
        node = qn.value
        if node in visited:
            continue
        visited.add(node)
        order.append(node)
        if node == goal:
            p = []
            x = goal
            while x is not None:
                p.append(x)
                x = parent.get(x)
            p.reverse()
            return order, p
        for nb in graph.get(node, []):
            if nb not in visited:
                if nb not in parent:
                    parent[nb] = node
                q.append(QNode(nb))
    return order, []

print("DLS / BFS Demo")
s = input("Start (default A): ").strip() or "A"
g = input("Goal (default M): ").strip() or "M"
try:
    dlim = int(input("Depth limit for DLS (default 3): ").strip() or "3")
except:
    dlim = 3

path = []
ok = dls(s, g, 0, dlim, path)
print("\nDLS found:", ok, " path:", path if ok else "[]")

order1, p1 = bfs_without_queue(s, g)
print("\nBFS(no queue/node) order:", order1)
print("BFS(no queue/node) path:", p1)

order2, p2 = bfs_with_queue(s, g)
print("\nBFS(with queue & node) order:", order2)
print("BFS(with queue & node) path:", p2)
```
